```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox



def predict_price():
   # window = tk.Tk()
    window.title("Price Laptob")
    window.config(background="pink")
    laptop_type = laptop_type_entry.get()  # الحصول على نوع اللابتوب من مربع النص
    if laptop_type == "lenovo":
        # الكود المرتبط بتنبؤ سعر نوع 1
        price = "1500"
    elif laptop_type == "Avita":
        # الكود المرتبط بتنبؤ سعر نوع 2
        price = "700"
    elif laptop_type == "ASUS":
        # الكود المرتبط بتنبؤ سعر نوع 3
        price = "500"
    elif  laptop_type == "SAMSUNG":
        # الكود المرتبط بتنبؤ سعر نوع 1
        price = "1900"
    elif laptop_type == "Hp":
        # الكود المرتبط بتنبؤ سعر نوع 2
        price = "900"
    elif laptop_type == "acer":
        # الكود المرتبط بتنبؤ سعر نوع 3
        price = "800"
    elif laptop_type == "Dell":
            # الكود المرتبط بتنبؤ سعر نوع 2
        price = "100"
    elif laptop_type == "RedmiBook":
            # الكود المرتبط بتنبؤ سعر نوع 3
        price = "400"
    else:
        messagebox.showerror("خطأ", "الرجاء اختيار نوع صحيح للابتوب")
        return

    messagebox.showinfo("تنبؤ بسعر اللابتوب", f"سعر اللابتوب المتوقع: ${price}")

#def handle_selection(event):
    selected_item = combobox.get()
    print(f"تم اختيار: {selected_item}")

window = tk.Tk()
window.title("تنبؤ بسعر اللابتوب")

laptop_type_label = tk.Label(window, text="نوع اللابتوب:")
laptop_type_label.pack()

laptop_type_entry = tk.Entry(window)
laptop_type_entry.pack()

# إنشاء نافذة التطبيق

# إنشاء قائمة الخيارات
options = ["Lenovo ", "Avita", "ASUS","SAMSUNG","Hp","acer","Dell","RedmiBook"]

# إنشاء عنصر Combobox
combobox = ttk.Combobox(window, values=options)
combobox.pack()

# os_bit
options1 = [" 64", "32"]
combobox1 = ttk.Combobox(window, values=options1)
combobox1.pack()

# ram
options2 = [" 4", "8","16","32"]
combobox2 = ttk.Combobox(window, values=options2)
combobox2.pack()

# os النظام
options3 = ["Mac", "Dos","Windows"]
combobox3 = ttk.Combobox(window, values=options3)
combobox3.pack()

options4 = ["0", "1","2","3"]
combobox4 = ttk.Combobox(window, values=options4)
combobox4.pack()

options5 = ["Yes", "No"]
combobox5 = ttk.Combobox(window, values=options5)
combobox5.pack()

# تعيين دالة التعامل مع الاختيار
combobox.bind("<<ComboboxSelected>>", predict_price)

button = tk.Button(window, text="تنبؤ بالسعر", command=predict_price)
button.pack()

window.mainloop()
```
